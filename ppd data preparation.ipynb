```python
import pandas as pd
from pathlib import Path
my_folder = Path('C:/Essex Msc/Project Proposal/Proposal data/ppd_data/').rglob('*.csv')
files = [x for x in my_folder]
files
```

```python
print(len(files))
```

```python
fields=['unique_id','price_paid','deed_date','postcode','property_type','new_build','estate_type','saon','paon','street','locality','town', 'district', 'county','transaction_category'  ]
dfs = [pd.read_csv(csv_file, usecols=fields) for csv_file in files]
```

```python
results = pd.concat(dfs)
display(results)
```

```python
results['unique_transact'] =results['unique_id'] +results['deed_date'].str[0:2]+results['deed_date'].str[3:5]+results['deed_date'].str[6:10]
```

```python
results_amend1=results.dropna(subset=['postcode'])
print(results_amend1)
```

```python
# splitting dataframe by row index 
#(data is split into a set of csv files under 10MB each, so each can be loaded into HarperDB -10MB is the limitation for a single file load)
myRows=int(round(results_amend1.shape[0]/50,0))
print(myRows)
allRows=results_amend1.shape[0]
print(allRows)
fromRow=0 
toRow=myRows
dfOutdfOut= results_amend1.iloc[:toRow,:]
dfOutdfOut.to_csv('C:/Essex Msc/Project Proposal/Proposal data/ppd_data/ppdresults_' +  "{}.csv".format(toRow), index=False)
while toRow<=allRows:
  fromRow=toRow+1  
  toRow=fromRow+ myRows
  dfOut = results_amend1.iloc[fromRow:toRow,:]  
  dfOut.to_csv('C:/Essex Msc/Project Proposal/Proposal data/ppd_data/ppdresults_' +  "{}.csv".format(toRow), index=False)
```
